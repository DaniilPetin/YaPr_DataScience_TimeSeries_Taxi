#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

План работы:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

## Обучение

### Создание признаков

```python
def make_features(data, max_lag, rolling_mean_size):
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)
    
    data.dropna(inplace=True)
    
    return data
```

```python
taxi_learning = taxi.copy()
taxi_learning = make_features(taxi_learning, 170, 12)

display(taxi_learning.head())
```

```python
taxi_learning.shape
```

### Разделение данных на выборки

```python
RANDOM_STATE = 42

X = taxi_learning.drop(columns='num_orders', axis=1) # входные признаки
y = taxi_learning['num_orders'] # целевой признак

X_train, X_test, y_train, y_test = train_test_split(
    X, y, shuffle=False,
    test_size=0.1 
)
```

### LinearRegression

```python
rmse_scorer = make_scorer(mean_squared_error, squared=False, greater_is_better=False)
```

```python
tscv = TimeSeriesSplit(n_splits=5) # кросс-валидация для временных рядов
```

```python
# инициализация модели линейной регрессии
model_lr  = LinearRegression()
```

```python
# предсказания модели
cv_scores_lr = cross_val_score(model_lr, X_train, y_train, 
                           scoring=rmse_scorer, cv=tscv)
model_lr.fit(X_train, y_train)
```

```python
# расчет значения RMSE для линейной регрессии
rmse_lr = round(np.abs(np.mean(cv_scores_lr)), 2)
print('Среднее значение RMSE Линейной регрессии =', rmse_lr)
```

### KNeighborsRegressor

```python
param_grid_knr = {
    'n_neighbors': range(2, 10),
    'weights': ['uniform', 'distance'],
    'p': [1, 1.5, 2, 3],
    'n_jobs': [-1]
}
```

```python
grid_search_knr = GridSearchCV(
    estimator = KNeighborsRegressor(),
    param_grid = param_grid_knr,
    cv=tscv,
    scoring=rmse_scorer,
    n_jobs=-1,
)

grid_search_knr.fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', grid_search_knr.best_estimator_)
print ('Метрика RMSE лучшей модели на кросс-валидации:', np.abs(grid_search_knr.best_score_))
```

```python
model_knr = grid_search_knr.best_estimator_
rmse_knr = np.abs(grid_search_knr.best_score_)
```

### DecisionTreeRegressor

```python
param_grid_dtr = {
    'max_depth': range(2, 15),
    'max_features': ['auto', 'sqrt', 'log2'] + list(range(2, 180, 10)),
    'min_samples_split': range(2, 15),
    'min_samples_leaf': range(1, 21),
    'max_leaf_nodes': [None] + list(range(10, 201, 20))
}
```

```python
randomized_search_dtr = RandomizedSearchCV(
    estimator = DecisionTreeRegressor(random_state=RANDOM_STATE),
    param_distributions = param_grid_dtr,
    n_iter=10,
    cv=tscv,
    scoring=rmse_scorer,
    n_jobs=-1,
    random_state=RANDOM_STATE
)

randomized_search_dtr.fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', randomized_search_dtr.best_estimator_)
print ('Метрика RMSE лучшей модели на кросс-валидации:', np.abs(randomized_search_dtr.best_score_))
```

```python
model_dtr = randomized_search_dtr.best_estimator_
rmse_dtr = np.abs(randomized_search_dtr.best_score_)
```

### RandomForestRegressor

```python
param_grid_rfr = {
    'n_estimators': [50, 100, 200, 300, 500],
    'max_depth': [None, 5, 10, 15, 20, 30],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 8],
    'max_features': ['auto', 'sqrt', 'log2', 0.3, 0.5, 0.7],
    'max_leaf_nodes': [None, 50, 100, 200],
}
```

```python
randomized_search_rfr = RandomizedSearchCV(
    estimator = RandomForestRegressor(random_state=RANDOM_STATE),
    param_distributions = param_grid_dtr,
    n_iter=10,
    cv=tscv,
    scoring=rmse_scorer,
    n_jobs=-1,
    random_state=RANDOM_STATE
)

randomized_search_rfr.fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', randomized_search_rfr.best_estimator_)
print ('Метрика RMSE лучшей модели на кросс-валидации:', np.abs(randomized_search_rfr.best_score_))
```

```python
model_rfr = randomized_search_rfr.best_estimator_
rmse_rfr = np.abs(randomized_search_rfr.best_score_)
```

### LightGBM

```python
param_grid_gbm = {
    'num_leaves': [25, 50],              
    'max_depth': [3, 5, -1],                 
    'min_data_in_leaf': [20, 50],  
    'learning_rate': [0.05, 0.1],
    'n_estimators': [100, 200],
    'feature_fraction': [0.7, 0.9],
    'bagging_freq': [1, 3],
}
```

```python
randomized_search_gbm = RandomizedSearchCV(
    estimator=LGBMRegressor(random_state=42, verbosity=-1),
    param_distributions=param_grid_gbm,
    n_iter=7,
    cv=tscv,
    scoring=rmse_scorer,
    n_jobs=-1,
    verbose=0,
    random_state=RANDOM_STATE,
    error_score='raise'
)

randomized_search_gbm.fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', randomized_search_gbm.best_estimator_)
print ('Метрика RMSE лучшей модели на кросс-валидации:', np.abs(randomized_search_gbm.best_score_))
```

```python
model_gbm = randomized_search_gbm.best_estimator_
rmse_gbm = np.abs(randomized_search_gbm.best_score_)
```

### Сравнение моделей

```python
compare = pd.DataFrame(
    index = ['LinearRegression', 'KNeighborsRegressor', 'DecisionTreeRegressor',
             'RandomForestRegressor', 'LightGBM'],
    columns = ['RMSE модели']
)
compare['RMSE модели'] = rmse_lr, rmse_knr, rmse_dtr, rmse_rfr, rmse_gbm

display(compare)
```

### Вывод по шагу 3

- Были протестированы модели `LinearRegression`, `KNeighborsRegressor`, `DecisionTreeRegressor`, `RandomForestRegressor`, `LightGBM`;
- Для каждой модели подобраны гиперпараметры при помощи кросс-валидации;
- Лучшую метрику RMSE на кросс-валидации показала модель `LightGBM`. Ee *RMSE=22.6*;
- Хуже всех с задачей справилась модель `DecisionTreeRegressor` - *RMSE=26.75*;
- Лучшая модель будет использована для предсказания на тестовой выборке.

## Тестирование

### Предсказание на тестовых данных

```python
model_best = model_gbm
```

```python
best_predictions = model_best.predict(X_test)
```

```python
rmse_test = mean_squared_error(y_test, best_predictions, squared = False)
print('RMSE модели LightGBM на тестовых данных =', round(rmse_test, 2))
```

### Важность признаков
